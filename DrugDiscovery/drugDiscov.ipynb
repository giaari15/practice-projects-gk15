```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df.head(3)
```

```python
df.standard_type.unique()
```

Find particular standard_type, in this case IC50, to uniformly analyze the potency of a drug. IC50 is the half-maximal inhibitory concentration, meaning the concentration of drug needed to inhibit a biological process or response by 50%. The lower the number of standard_value (concentration of the drug) of the drug, the better the potency. The higher the number, the greater the amount of the drug is needed to produce the same inhibition at 50%.

```python
df.to_csv('bioactivity_data.csv', index=False)
```

```python
df2 = df[df.standard_value.notna()]
df2.head(3)
```

This bioactivity data is in the IC50 unit. Compounds having less than 1000nM will be considered active. Compounds having values more than 10,000 nM will be considered inactive. Values in between 1000 and 10,000 nM will be referred to as intermediate.

```python
bioactivity_class = pd.Series(name="bioactivity")
for i, val in enumerate(df2.standard_value):
    if float(val) >= 10000:
        bioactivity_class[i] = "inactive"
    elif float(val) <= 1000:
        bioactivity_class[i] = "active"
    else:
        bioactivity_class[i] = "intermediate"
bioactivity_class
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3
```

```python
df3_1 = df3.reset_index(drop=True)
```

```python
df4 = pd.concat([df3_1, bioactivity_class], axis = 1)
df4
```

```python
df4.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
